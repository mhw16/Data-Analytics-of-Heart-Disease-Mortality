```python
#Importing the necessary python packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the file 
data = pd.read_csv('HeartMortality.csv')
print(data)
```

```python
#Printing the statistical summary of 'HeartMortality'
data.describe()
```

```python
#Identifying null values
data.isnull()
```

```python
#Cleaning two of the columns of NaN values, since they are footnotes the values do not necessarily matter
cleaned=data.fillna({'Data_Value_Footnote_Symbol':0,'Data_Value_Footnote':0})
print(cleaned)
```

```python
#Cleaning the rows of values that do not have a number per 100,000.
#I decided to drop the rows because it would be wrong to input falsified statistics.
#This instance is not permanant.
cleaned.dropna()
```

```python
#Checking the missing values
cleaned.isnull()
```

```python
#Grouping important aspects of data (Location, Gender, and ethnicity)
state = cleaned.groupby('LocationAbbr').mean()
print(state)
gender = cleaned.groupby('Stratification1').mean()
print(gender)
race = cleaned.groupby('Stratification2').mean()
print(race)
date = cleaned.groupby('Year').mean()
print(date)
```

```python
#Initial Graphical Summary
import matplotlib.pyplot as plt
%matplotlib inline

gender['Data_Value'].plot.bar()
```

```python
race['Data_Value'].plot.bar()
```

```python
import seaborn as sns; sns.set(color_codes=True)
sns.kdeplot(cleaned.Data_Value[cleaned.Stratification1 =='Male'], label ='men', shade = True)
sns.kdeplot(cleaned.Data_Value[cleaned.Stratification1 =='Female'], label ='women', shade = True)
plt.xlabel('Heart Disease Mortality')
```
